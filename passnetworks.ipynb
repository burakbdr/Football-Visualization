## Imports

```python
import pandas as pd
from mplsoccer.pitch import Pitch
```

## Get Data 

```python
df = pd.read_csv('/Users/burakbedir/Desktop/B/Projects/Football-Visualization/datasets/valladolidA.csv')

df
```

```python
# Filter df to get only Barcelona
df = df[df['teamId']== 'Barcelona']
```

```python
df
```

```python
# Create passer and reciever columns
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1)

# Filter df only for passes 
passes = df[df['type']=='Pass']
succesfull = passes[passes['outcome']=='Successful']
```

```python
succesfull
```

```python
subs = df[df['type'] == 'SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()
```

```python
firstSub
```

```python
succesfull = succesfull[succesfull['minute'] < firstSub]
```

```python
succesfull
```

```python
# Change dtype of pass and recvr
passr= pd.to_numeric(succesfull['passer'], downcast='integer')
rec= pd.to_numeric(succesfull['recipient'], downcast='integer')

succesfull['passer'] = passr
succesfull['recipient'] = rec
```

```python
succesfull
```

```python
# Create average locations
average_locations = succesfull.groupby('passer').agg({'x':['mean'], 'y':['mean','count']}) 
average_locations.columns = ['x','y','count']
```

```python
average_locations
```

```python
pass_between = succesfull.groupby(['passer','recipient']).id.count().reset_index()
pass_between.rename({'id':'pass_count'}, axis='columns',inplace=True)

#merge to get all new df
pass_between = pass_between.merge(average_locations, left_on='passer', right_index=True)
pass_between = pass_between.merge(average_locations, left_on='recipient', right_index=True,suffixes=['','_end'])
```

```python
pass_between
```

```python
# High passing team like Barcelona we can filter 
pass_between = pass_between[pass_between['pass_count']>3]
```

```python
pass_between
```

```python
# Plot the pitch

pitch = Pitch(pitch_type='uefa', pitch_color='#0583F2', line_color='#FFFCF7')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#FFFCF7')


# Plot the arrows

arrows = pitch.arrows(1.2*pass_between.x,.8*pass_between.y,1.2*pass_between.x_end,.8*pass_between.y_end,
                     width = 5, headwidth = 5, color = '#D90479', ax = ax, zorder = 1, alpha = .5)

# Plot the nodes

nodes = pitch.scatter(1.2*average_locations.x,.8*average_locations.y,
                     s = 300, color = '#FFFCF7', edgecolors = 'black', linewidth = 2.5, alpha = 1, zorder = 1, ax=ax)
```
